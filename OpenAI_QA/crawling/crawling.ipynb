```python
# jobCategory 별로 정보 저장
keys = ('jobCategories', 'responsibility', 'qualifications', 'preferredRequirements')

position_statistics = {}
for position in positions:
    id = position['jobCategories_id']
    if id in position_statistics:
        position_statistics[id]['count'] += 1
        for key in keys:
            position_statistics[id][key].append(position[key])
        for stack in position['techStacks']:
            if stack.lower() in position_statistics[id]['techStacks']:
                position_statistics[id]['techStacks'][stack.lower()] += 1
            else:
                position_statistics[id]['techStacks'][stack.lower()] = 1
    else:
        position_statistics[id] = {key: [position[key]] for key in keys}
        position_statistics[id]['count'] = 1
        position_statistics[id]['techStacks'] = {stack.lower(): 1 for stack in position['techStacks']}

for position in position_statistics:
    position_statistics[position]['jobCategories'] = position_statistics[position]['jobCategories'][0]

position_statistics_df = pd.DataFrame(position_statistics).transpose()
position_statistics_df.to_csv("position_statistics.csv", index=False)
```

```python
# 직무별 요구하는 프로그래밍 언어 비율 구하기

from collections import Counter
total = len(positions)
for position in position_statistics:
    portion = round(100 * position_statistics[position]['count'] / total, 2)
    stacks = Counter(position_statistics[position]['techStacks']).most_common(3)
    counts = len(position_statistics[position]['responsibility'])
    top_3 = {stacks[i][0]: round(100*stacks[i][1]/counts, 2) for i in range(3)}
    print(position_statistics[position]['jobCategories'], portion, top_3)


# for position in position_statistics:
#     m, stack = 0, ''
#     for k, v in position_statistics[position]['techStacks'].items():
#         if v > m:
#             stack , m = k, v
#     print(position_statistics[position]['jobCategories'], stack, round((m/len(position_statistics[position]['responsibility']))*100, 2))
```

```python
from collections import Counter
total = len(positions)
stacks_portion = {}
for position in position_statistics:
    portion = round(100 * position_statistics[position]['count'] / total, 2)
    stacks = Counter(position_statistics[position]['techStacks'])
    counts = len(position_statistics[position]['responsibility'])
    stacks = {s: round(c/counts*portion, 2) for s, c in stacks.items()}
    for stack in stacks:
        if stack in stacks_portion:
            stacks_portion[stack] += stacks[stack]
        else:
            stacks_portion[stack] = stacks[stack]
stacks_portion
```
